```python
# Loading in all packages from the jupyter environment
import numpy as np
import os
import pandas as pd
from itertools import islice
import re

# command for splitting the GM5.0 file into multiple files: 
#awk 'BEGIN{x=1} NR>1 && /^>GM/{ close("file"x); x++ }{ print > ("file"x) }' /home/julian/rgtdata/motifs/gimme/gimme.txt
# env: jupyter
```

```python
# counting number of lines to slice
#print(os.listdir("/home/julian/rgtdata/motifs/gimme_converted/"))

for i in os.listdir("/home/julian/rgtdata/motifs/gimme_converted/"):
    with open('/home/julian/rgtdata/motifs/gimme_converted/'+ i) as lines:
        num_lines = sum(1 for line in lines)
        #print(num_lines)
    
    # load in the array and transpose it (acgt columns to acgt as rows)
    with open('/home/julian/rgtdata/motifs/gimme_converted/'+ i) as lines:
        array = np.genfromtxt(islice(lines, 1, num_lines))
        array = np.transpose(array)
        #print(array)

    # load the header as the file name
    with open('/home/julian/rgtdata/motifs/gimme_converted/'+ i) as lines:
        header = lines.readline()
        header = header.replace(">", "")
        header = header.replace("\t*", "")
        #print(header)
        if not header.startswith('#'):
            out_name = "/home/julian/rgtdata/motifs/gimme_converted2/"+str(header)+".pwm"
            
            #remove apparent newline character from the out_name
            pattern = "[\n|\r|\r\n]"
            out_name = re.sub(pattern, '', out_name )
            print(out_name)
            np.savetxt(out_name, array, fmt='%.18e', delimiter=' ', newline='\n', header='', footer='', comments='#', encoding=None)
        
```
